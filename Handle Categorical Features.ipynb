## KDD Tech
This involves taking only those of the categories in a feature which have high frequency.
The no of such categories varies as per the feature : Check the value count of each of the category to decide the threshold

```python
lst_10 = df.X1.value_counts().sort_values(ascending = False).head(10).index
```

```python
lst_10 = list(lst_10)
lst_10
```

## 5. Target Guided Ordinal Encoding

1.Ordering the labels according to the target

2.Replace the labels by the joint probability of being 1 or 0

```python
df = pd.read_html("https://github.com/krishnaik06/Feature-Engineering-Live-sessions/blob/master/titanic.csv")
df = df[0]
df = df.loc[:,["Cabin","Survived"]]
df.head()
```

```python
df.isnull().sum()
```

```python
df.Cabin.fillna('Missing',inplace = True)
```

```python
df.head()
```

```python
df["Cabin"].astype(str).str[0]
```

```python
df["Cabin"] = df["Cabin"].astype(str).str[0]
```

```python
df.head()
```

```python
df.Cabin.unique()
```

```python
df.groupby(['Cabin'])['Survived'].mean().sort_values()
```

```python
Cabin_ordinal = df.groupby(['Cabin'])['Survived'].mean().sort_values().index
```

```python
Cabin_ordinal
```

```python
ord_lab = {k:i for i,k in enumerate(Cabin_ordinal,0)}
ord_lab
```

```python
df["Cabin_Ordinal_Labels"] = df["Cabin"].map(ord_lab)
df.head()
```

## 7. Probability Ratio Encoding

```python
df = pd.read_csv("titanic.csv", usecols = ['Cabin',"Survived"])
df.head()
```

```python
df["Cabin"].fillna("Missing", inplace = True)
```

```python
df["Cabin"].unique()
```

```python
df["Cabin"] = df["Cabin"].astype(str).str[0]
df.head()
```

```python
df.Cabin.unique()
```

```python
prob_df = df.groupby(['Cabin'])["Survived"].mean()
```

```python
prob_df = pd.DataFrame(prob_df)
```

```python
prob_df
```

```python
prob_df['Died'] = 1-prob_df['Survived']
prob_df
```

```python
prob_df["Probability_Ratio"] = prob_df['Survived']/prob_df['Died']
prob_df
```

```python
prob_encoded = prob_df["Probability_Ratio"].to_dict()
```

```python
df["Cabin_Prob_Encod"] = df["Cabin"].map(prob_encoded)
df.head()
```
